## Extracción, Transformación y Limpieza de los datos

* Observamos los valores nulos

* Filas duplicadas

```python
tablas = [MSFT_balance_general, MSFT_estado_resultados, Alphabet_balance_general, Alphabet_estado_resultados]
nombre_tabla = ['MSFT_balance_general', 'MSFT_estado_resultados', 'Alphabet_balance_general', 'Alphabet_estado_resultados']
for i in range(len(tablas)):
    print(f"------- {nombre_tabla[i]} --------")
    print()
    print(tablas[i].duplicated().sum())
    print(f"----------------------------------")
    print(f"----------------------------------")
    print()
```

```python
tablas = [MSFT_balance_general, MSFT_estado_resultados, Alphabet_balance_general, Alphabet_estado_resultados]

for k in tablas:
    columnas = k.columns.to_list()

    for i in range(len(columnas)):
        if columnas[i] == columnas[0]:
            None
        else:
           k[columnas[i]]= k[columnas[i]].apply(lambda x: str(x).replace('.',''). replace(',','.'))
    
    for i in range(1, k.shape[1]):
        for j in range(k.shape[0]):
            valor = k.iloc[j, i]

            if "." in valor:
                k.iloc[j, i] = float(valor)
            elif "--" in valor:
                k.iloc[j, i] = str(valor)
            else: 
                k.iloc[j, i] = int(valor)
```

Pasar todos los números a millones

```python
tablas = [MSFT_balance_general, MSFT_estado_resultados, Alphabet_balance_general, Alphabet_estado_resultados]

for k in tablas:
    
    for i in range(1, k.shape[1]):
        
        for j in range(k.shape[0]):
            valor = k.iloc[j, i]

            if valor == "--":
                k.iloc[j, i] = valor
            else: 
                k.iloc[j, i] = valor/1000
```
