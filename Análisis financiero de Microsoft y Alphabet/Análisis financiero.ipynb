## Extracción, Transformación y Limpieza de los datos

* Observamos los valores nulos

```python
tablas = [MSFT_balance_general, MSFT_estado_resultados, Alphabet_balance_general, Alphabet_estado_resultados]
nombre_tabla = ['MSFT_balance_general', 'MSFT_estado_resultados', 'Alphabet_balance_general', 'Alphabet_estado_resultados']
for i in range(len(tablas)):
    print(f"------- {nombre_tabla[i]} --------")
    print()
    print(tablas[i].isnull().sum())
    print(f"----------------------------------")
    print(f"----------------------------------")
    print()
    
```

* Filas duplicadas

```python
tablas = [MSFT_balance_general, MSFT_estado_resultados, Alphabet_balance_general, Alphabet_estado_resultados]
nombre_tabla = ['MSFT_balance_general', 'MSFT_estado_resultados', 'Alphabet_balance_general', 'Alphabet_estado_resultados']
for i in range(len(tablas)):
    print(f"------- {nombre_tabla[i]} --------")
    print()
    print(tablas[i].duplicated().sum())
    print(f"----------------------------------")
    print(f"----------------------------------")
    print()
```

```python
tablas = [MSFT_balance_general, MSFT_estado_resultados, Alphabet_balance_general, Alphabet_estado_resultados]

for k in tablas:
    columnas = k.columns.to_list()

    for i in range(len(columnas)):
        if columnas[i] == columnas[0]:
            None
        else:
           k[columnas[i]]= k[columnas[i]].apply(lambda x: str(x).replace('.',''). replace(',','.'))
    
    for i in range(1, k.shape[1]):
        for j in range(k.shape[0]):
            valor = k.iloc[j, i]

            if "." in valor:
                k.iloc[j, i] = float(valor)
            elif "--" in valor:
                k.iloc[j, i] = str(valor)
            else: 
                k.iloc[j, i] = int(valor)
```

Pasar todos los números a millones

```python
tablas = [MSFT_balance_general, MSFT_estado_resultados, Alphabet_balance_general, Alphabet_estado_resultados]

for k in tablas:
    
    for i in range(1, k.shape[1]):
        
        for j in range(k.shape[0]):
            valor = k.iloc[j, i]

            if valor == "--":
                k.iloc[j, i] = valor
            else: 
                k.iloc[j, i] = valor/1000
```

### Ratios de Liquidez

* Liquidez corriente

```python
# Alphabet_balance_general.iloc[1,:] # Activo corriente
# Alphabet_balance_general.iloc[33,:] # Pasivo corriente
```

```python
# MSFT_balance_general.iloc[1,:] # Activo corriente
# MSFT_balance_general.iloc[37,:] # Pasivo corriente
```

```python
columnas = MSFT_balance_general.columns.to_list()[5:0:-1]

liquidez_corriente_MSFT = []
for i in columnas:
        lc_msft = round(MSFT_balance_general.loc[1,i]/ MSFT_balance_general.loc[37,i],1)
        liquidez_corriente_MSFT.append(lc_msft)

print(liquidez_corriente_MSFT)
```

```python
columnas = Alphabet_estado_resultados.columns.to_list()[5:1:-1]

liquidez_corriente_alphabet = []
for i in columnas:
    lc_alphabet = round(Alphabet_balance_general.loc[1,i]/ Alphabet_balance_general.loc[33,i],1)
    liquidez_corriente_alphabet.append(lc_alphabet)

print(liquidez_corriente_alphabet)
```

```python
liquidez_corriente = pd.DataFrame([liquidez_corriente_alphabet, liquidez_corriente_MSFT], index=['Alphabet', 'Microsoft'], columns=[Alphabet_estado_resultados.columns.to_list()[5:1:-1]])
liquidez_corriente
```

```python
columnas = MSFT_balance_general.columns.to_list()[5:0:-1]

capital_trabajo_MSFT = []
for i in columnas:
        lc_msft = MSFT_balance_general.loc[1,i] - MSFT_balance_general.loc[37,i]
        capital_trabajo_MSFT.append(lc_msft)

print(capital_trabajo_MSFT)
```

```python
capital_de_trabajo = pd.DataFrame([capital_trabajo_alphabet, capital_trabajo_MSFT], index=['Alphabet', 'Microsoft'], columns=[Alphabet_estado_resultados.columns.to_list()[5:1:-1]])
capital_de_trabajo
```

```python
columnas = Alphabet_estado_resultados.columns.to_list()[5:1:-1]

ratio_endeudamiento_alphabet = []
for i in columnas:
    re_alphabet = round((Alphabet_balance_general.loc[33,i] +  Alphabet_balance_general.loc[47,i]) / Alphabet_balance_general.loc[57,i],2)
    ratio_endeudamiento_alphabet.append(re_alphabet)

print(ratio_endeudamiento_alphabet)
```

```python
columnas = MSFT_balance_general.columns.to_list()[5:0:-1]

ratio_endeudamiento_MSFT = []
for i in columnas:
        re_msft = round((MSFT_balance_general.loc[37,i] + MSFT_balance_general.loc[48,i]) / MSFT_balance_general.loc[52,i] ,2)
        ratio_endeudamiento_MSFT.append(re_msft)

print(ratio_endeudamiento_MSFT)
```

```python
ratio_de_endeudamiento = pd.DataFrame([ratio_endeudamiento_alphabet, ratio_endeudamiento_MSFT], index=['Alphabet', 'Microsoft'], columns=[Alphabet_estado_resultados.columns.to_list()[5:1:-1]])
ratio_de_endeudamiento
```

* Cobertura de Deuda Financiera

```python
# Alphabet_balance_general.loc[42, i]:
# Deuda corriente y obligaciones por arrendamiento de capital

# Alphabet_balance_general.loc[48, i]:
# Deuda a largo plazo y obligaciones por arrendamiento de capital

# Alphabet_estado_resultados.loc[39, i]:
# EBITDA

columnas = Alphabet_estado_resultados.columns.to_list()[5:1:-1]

cobertura_de_deuda_alphabet = []
for i in columnas:
    cf_alphabet = round((Alphabet_balance_general.loc[42,i] +  Alphabet_balance_general.loc[48,i]) / Alphabet_estado_resultados.loc[39,i],1)
    cobertura_de_deuda_alphabet.append(cf_alphabet)

print(cobertura_de_deuda_alphabet)
```

```python
# MSFT_balance_general.loc[43,i]:
# Deuda corriente
# MSFT_balance_general.loc[49,i]:
# Deuda a largo plazo
# MSFT_estado_resultados.loc[38,i]:
# EBITDA

columnas = MSFT_balance_general.columns.to_list()[5:0:-1]

cobertura_de_deuda_MSFT = []
for i in columnas:
        cf_msft = round((MSFT_balance_general.loc[43,i] + MSFT_balance_general.loc[49,i]) / MSFT_estado_resultados.loc[38,i] ,1)
        cobertura_de_deuda_MSFT.append(cf_msft)

print(cobertura_de_deuda_MSFT)
```

```python
Cobertura_de_Deuda_Financiera = pd.DataFrame([cobertura_de_deuda_alphabet, cobertura_de_deuda_MSFT], index=['Alphabet', 'Microsoft'], columns=[Alphabet_estado_resultados.columns.to_list()[5:1:-1]])
Cobertura_de_Deuda_Financiera
```
